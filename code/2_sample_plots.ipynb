## Data visualization script
To iterate through the data used in training.   

```python
import utils as ut 
import matplotlib.pyplot as plt
```

```python
def display_examples(dataset, num_examples=5):
    for i in range(num_examples):
        example = dataset[i]
        slice1 = example['slice1']
        slice2 = example['slice2']
        label = example['label']
        img_type = 'T1' if example['image_type'] == 0 else 'T2'

        plt.figure(figsize=(10, 5))

        # Slice 1
        plt.subplot(1, 2, 1)
        plt.imshow(slice1, cmap='gray')
        plt.title(f"Slice 1 | {img_type} | Label: {label}")
        plt.axis('off')

        # Slice 2
        plt.subplot(1, 2, 2)
        plt.imshow(slice2, cmap='gray')
        plt.title(f"Slice 2 | {img_type} | Label: {label}")
        plt.axis('off')

        plt.show()
```

```python
# Latest function to generate the folds

# Change noise size for visual display
NOISE_SIZE = 5
ds_folds = ut.siamese_noise_dataset_fold_range(noise_size=NOISE_SIZE)

first_fold = ds_folds[0]
# Train set 
train_ds = first_fold['train']
display_examples(train_ds, num_examples=4)
```
